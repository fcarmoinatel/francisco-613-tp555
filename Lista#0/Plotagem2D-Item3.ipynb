```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

#Número de elementos conforme especificado no item 3 da lista #0
M = 1000

# conforme especificado no item 3 da lista #0, teremos y = [1000,1]

x = np.random.rand(M) # Distribuição Aleatoria Uniforme
w = np.random.randn(M) # Distribuição Aleatória Gaussiana


# Equação proposta no Exercício 3 da Lista #0

y = 1.2 + 2.3*x + 10*w

#plotagem dos axis (x, y)
fig1, axis = plt.subplots()

#plotagem dos pontos no gráfico
plt.scatter(x, y, color = "green")

axis.set(xlabel='x', ylabel='y' , title='Equation: y = 1.2 + 2.3*x + 10*w')
plt.grid()

fig1.savefig("Plotagem2D-Item3.png")

plt.show()
```
