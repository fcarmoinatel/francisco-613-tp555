```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import axes3d # Importando base para grÃ¡ficos 3D conforme material de base descrito nas URLs abaixo.
import numpy as np

%matplotlib notebook

fig = plt.figure()
ax = fig.gca(projection='3d')

# Criando o arranjo com base na literatura proposta em: https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html

X1 = np.arange(-10, 11, 0.25)
X2 = np.arange(-10, 11, 0.25)
X1, X2 = np.meshgrid(X1, X2)
Y = X1**2 + X2**2

# Plotagem da superfÃ­cie 3D conforme literatura proposta em: https://matplotlib.org/3.1.3/gallery/mplot3d/surface3d.html

surf = ax.plot_surface(X1, X2, Y, cmap=cm.coolwarm,linewidth=0, antialiased=False)
ax.plot_wireframe(X1, X2, Y)


# Adicionando uma barra de cores para mapeamento
fig.colorbar(surf, shrink=0.5, aspect=10)

fig.savefig("Plotagem3D-Item5.png")
plt.show()
```
