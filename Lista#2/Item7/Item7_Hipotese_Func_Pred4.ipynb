```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

%matplotlib inline
```

```python
df = pd.read_csv('training.csv', header=None)
x_l = df[0].to_numpy()
y_l = df[1].to_numpy()
arr1 = [x_l]
arr2 = [y_l]
x = np.transpose(arr1)
y = np.transpose(arr2)
fig = plt.figure(figsize=(10,10))
plt.plot(x, y, 'b.')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Valores de aprendizado')
```

```python
M = 1000
np.random.seed(1234)
X = np.c_[np.ones((M, 1)), x, x**2, x**3, x**4]
a_opt = np.linalg.pinv(np.transpose(X).dot(X)).dot(np.transpose(X).dot(y))
yhat = a_opt[0, 0] + a_opt[1, 0]*x + a_opt[2, 0]*x**2 + a_opt[3, 0]*x**3 + a_opt[4, 0]*x**4
Joptimum = (1/M)*np.sum(np.power((y - yhat), 2))
```

```python
alpha = 0.00002
n_iterations = 1000
a = np.zeros((5,1))
a[0] = -10;
a[1] = -10;
a[2] = -10;
a[3] = -10;
a[4] = -10;
Jgd = np.zeros(n_iterations+1)
Jgd[0] = (1/M)*sum(np.power(y - X.dot(a), 2))

iteration = 0

while iteration < n_iterations:
    gradients = -2/M * X.T.dot(y - X.dot(a))
   
    a = a - alpha * gradients
   
    Jgd[iteration+1] = (1/M)*sum(np.power( (y - X.dot(a)) , 2))
    iteration = iteration + 1
```
