```python
# Import all necessary libraries.
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.linear_model.logistic import LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
```

```python
# Load digit data set.
digits = load_digits()
x = digits.data
y = digits.target
```

```python
# Display the first digit.
plt.figure(1, figsize=(3, 3))
plt.imshow(digits.images[-1], cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```

```python
# Split the data set.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Instantiate LogisticRegression object.
model = LogisticRegression(solver='lbfgs', max_iter=10000, multi_class='multinomial')
```

```python
# Train model.
model.fit(x_train, y_train)
```

```python
# Predict.
y_pred = model.predict(x_test)
```

```python
print("Accuracy of Logistic Regression model is:", accuracy_score(y_test, y_pred)*100)
```

```python
# Evaluate the model's performance with the confusion matrix between the true and predicted labels for the test data.
fig = plt.figure(figsize=(7,7))
mat = confusion_matrix(y_test, y_pred)
sns.set(font_scale=1.8)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, xticklabels=digits.target_names, yticklabels=digits.target_names, cmap="Blues")
plt.xlabel('True label', fontsize=24)
plt.ylabel('Predicted label', fontsize=24)
# Save figure
plt.savefig("confusion_matrix_digit_classification_softmax.png", dpi=600)
plt.show()
```
